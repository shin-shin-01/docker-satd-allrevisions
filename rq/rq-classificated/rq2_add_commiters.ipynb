```python
cnt_contributors = []

for idx, data_row in data.iterrows():
    # 分類後データ に対応する result を取得
    row = result.loc[idx, :]
    latestFilename = row["LatestDockerfile"]
    project = row["project"]
    deleteComment = row["DeleteComment"]
    # CSV ファイル名
    csv = "_".join(project.split()) + ".csv"
    
    # rename List を取得
    satd_add_info = pd.read_csv(f'../../6-satdCommentsAddInfo/{csv}', index_col=0)
    # 8 で日付変換しているためそれに合わせて取得（８で強引に日付を直してるのがよくない: JSTをUTCと呼んでる）
    # TODO :上記修正
    satd_add_info["Date"] = pd.to_datetime(satd_add_info["Date"], utc=True)
    satd_add_info["Date"] = satd_add_info["Date"].dt.tz_convert('Asia/Tokyo')
    satd_add_info["Date"] = satd_add_info["Date"].dt.strftime('%Y/%m/%d %H:%M')
    satd_add_info["Date"] = pd.to_datetime(satd_add_info["Date"], utc=True)
    
    satd_add_info = satd_add_info[ (satd_add_info["LatestDockerfile"] == latestFilename) & (satd_add_info["Date"] <= deleteComment)]
    
    # renameList ユニーク
    renameList = []
    for lst in satd_add_info["RenameList"]:
        renameList.extend(lst.split())
    renameList = list(set(renameList))
    
    # github のログから対応するコミット：コミット者を取得
    gitlog = pd.read_csv(f'../../2-gitlogsCSV/{csv}', index_col=0)
    gitlog["Date"] = pd.to_datetime(gitlog["Date"], utc=True)
    gitlog["Date"] = gitlog["Date"].dt.tz_convert('Asia/Tokyo')
    gitlog["Date"] = gitlog["Date"].dt.strftime('%Y/%m/%d %H:%M')
    gitlog["Date"] = pd.to_datetime(gitlog["Date"], utc=True)
    
    gitlog = gitlog[
        (gitlog["Dockerfile"].isin(renameList)) &
        (gitlog["Date"] <= deleteComment) &
        (gitlog["Merged?"] == False)
    ]
    
    # 名前が変わっているかもしれないし，メールが変わっているかもしれない
    aut_pair = []
    already_commit = []
    added_pair = []
    cnt = 0
    # Author が複数人いることに対応
    for auts in gitlog["Author"]:
        for aut in auts.splitlines():
            name = "/".join((aut.split("<")[0]).split())
            mail = aut.split("<")[1].split(">")[0]
            aut_pair.append([name, mail])

    for pair in aut_pair:
        if pair[1] == "noreply@github.com":
            continue
        if (not pair[0] in already_commit):
            already_commit.append(pair[0])
#             already_commit.append(pair[1])
            added_pair.append([pair[0], pair[1]])
            cnt += 1
    
    contributors = data_row[1]
    
    # SAME??
    before_name = "/".join((data_row[5].split("<")[0]).split())
    before_mail = data_row[5].split("<")[1].split(">")[0]
    after_name = "/".join((data_row[6].split("<")[0]).split())
    after_mail = data_row[6].split("<")[1].split(">")[0]
    
    is_same = "SAME" if (before_name == after_name) or (before_mail == after_mail) else "NOT"
    
    print(f'{idx:04}: renameList-{len(renameList)} : {contributors} → {cnt}', is_same)
    cnt_contributors.append(cnt)
    
#     if idx == 120:
#         print(added_pair)
#         display(gitlog)

#         break
```
